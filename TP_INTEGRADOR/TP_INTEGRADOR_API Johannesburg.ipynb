```python
import pandas as pd
import io
```

```python
import requests


url = "https://archive-api.open-meteo.com/v1/archive?latitude=-26.20&longitude=28.04&start_date=1940-01-01&end_date=2023-02-10&daily=temperature_2m_max,temperature_2m_min,temperature_2m_mean,precipitation_sum&timezone=Africa%2FJohannesburg"

response = requests.request("GET", url)


```

```python
json2= response.json()
```

```python
json2.keys()
```

```python
df_johannesburg = pd.DataFrame(json2['daily'])
```

```python
df_johannesburg
```

```python
df_johannesburg1 = df_johannesburg.loc[:"temperature_2m_max",:"temperature_2m_min"]

df_johannesburg1.rename(columns={"time":"Fecha",
                                 "temperature_2m_max":"temp_max",
                                 "temperature_2m_min":"temp_min"},inplace=True)

df_johannesburg1.dropna(inplace=True)

df_johannesburg1.to_csv('johannesburg.csv')
```
