```python
import pandas as pd
import io
```

```python
import requests



url = "https://archive-api.open-meteo.com/v1/archive?latitude=40.73&longitude=-73.93&start_date=1940-01-01&end_date=2023-02-10&daily=temperature_2m_max,temperature_2m_min,temperature_2m_mean,precipitation_sum&timezone=America%2FNew_York"

response = requests.request("GET", url)


```

```python
json2= response.json()
```

```python
json2.keys()
```

```python
df_ny = pd.DataFrame(json2['daily'])
df_ny
```

```python
df_ny1 = df_ny.loc[:"temperature_2m_max",:"temperature_2m_min"]

df_ny1.rename(columns={"time":"Fecha",
                                 "temperature_2m_max":"temp_max",
                                 "temperature_2m_min":"temp_min"},inplace=True)

df_ny1.dropna(inplace=True)

df_ny1.to_csv('new_york.csv')
```
